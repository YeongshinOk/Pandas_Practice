(1) 모듈 로딩 --------------------------------

```python
from pandas import DataFrame, read_csv
```

(2) 데이터 파일 준비 -------------------------------

```python
dataFile='../DATA/movies.csv'
```

(3) 데이터 저장 ---------------------------------

```python
movieDF = read_csv(dataFile)
```

```python
movieDF, type(movieDF)
```

(4) DataFrame의 속성들 출력

```python
movieDF.index, movieDF.columns, movieDF.values, movieDF.shape, movieDF.ndim, movieDF.dtypes
```

(5) DataFrame의 컬럼별로 따로 저장하기

```python
movieDF.columns
```

```python
Column_data = []

for idx, column in enumerate(movieDF.columns):
        print(idx, column)
        Column_data.append(movieDF[column])
        # 열 뽑을 때는 loc, iloc 쓰지 말고 그냥 출력
        # [] 속에 데이터 값 넣어야 함, 인덱스 넣으면 안됨
        print(Column_data[idx])
```

(6) 컬럼별 데이터들의 속성 즉, index, values, shape, ndim, dtype 정보 출력 

```python
for colName in movieDF.columns:
    print(f'--------------------------------------------{colName}의 속성 정보--------------------------------------------')
    print(f'index : {movieDF[colName].index}')
    print(f'values : {movieDF[colName].values}')
    print(f'shape : {movieDF[colName].shape}')
    print(f'ndim : {movieDF[colName].ndim}')
    print(f'size : {movieDF[colName].size}')
    print(f'dtype : {movieDF[colName].dtype}')
    print()
```

(7) 컬럼별 인덱스를 변경 실습

```python
movieDF = movieDF[['Title', 'Rank', 'Year', 'Studio', 'Gross']]

print(movieDF)
```

(8) 컬럼별 데이터 즉 값 변경

```python
movieDF.iloc[0,1] = 'merong!!'
movieDF
```
